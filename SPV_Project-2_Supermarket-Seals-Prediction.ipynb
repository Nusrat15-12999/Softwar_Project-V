````python
#Line Chart
plt.figure(figsize=(10,7))
plt.title('Line Chart of Product selling ',fontweight='bold')
plt.xlabel("Number of Products")
plt.ylabel("Cost_of_Goods_Sold")
Data['Cost_of_Goods_Sold'].plot()
````

````python
#Histogram
plt.figure(figsize=(10,7))
plt.title('Frequency of Products Unit_price ',fontweight='bold')
plt.xlabel("Unit Price Range")
Data['Unit_price(K)'].plot.hist(color=["Black"])
````

````python
#Horizontal simple bar charts
plt.figure(figsize=(8,5))

plt.barh(Data.Product_Category,Data.Profit)
plt.title('Store Inventory & Profit')
plt.ylabel('Product')
plt.xlabel('Profit')
plt.show()
````

````python
#Bar Chart
plt.figure(figsize=(10,7))
plt.title('Bar Chart of Product selling',fontweight='bold')

plt.xlabel("Number of Products",fontsize=20)
plt.ylabel("Total_Revenue(K)",fontsize=20)

Data["Total_Revenue(K)"].plot(kind="bar",color=["purple"])
````

````python
#scatter plot
Data["Total_Revenue(K)"]=Data["Total_Revenue(K)"].astype(int)
Data.plot(x="Profit",y="Total_Revenue(K)",kind="scatter")
````

````python
#Bar Plot
plt.figure(figsize=(10,7))

sns.barplot(Data['City'],Data['Profit'],palette='Set2')
plt.xlabel('City name',fontsize='16')
plt.xticks(fontsize='16')
plt.ylabel('Profit',fontsize='16')
plt.yticks(fontsize='16')
plt.suptitle("Most profittable city judgement\n",)
````

````python
#Pie Chart
plt.figure(figsize=(15,15))

plt.subplot(3,2,1)
x = Data['Branch'].value_counts()
y = Data['Branch'].unique()
c = ['b','y','r']
plt.pie(x,labels=y,colors=c,startangle=90,shadow=True)
plt.legend(loc=1)
plt.title("Supermarket branches",fontsize='20',fontweight='bold')

plt.subplot(3,2,2)
x = Data['City'].value_counts()
y = Data['City'].unique()
c = ['purple','green','cyan']
plt.pie(x,labels=y,colors=c,startangle=90,shadow=True)
plt.legend(loc=1)
plt.title("Cities of shops",fontsize='20',fontweight='bold')


plt.subplot(3,2,3)
x = Data['Payment'].value_counts()
y = Data['Payment'].unique()
c = ['#E34234','magenta','blue']
plt.pie(x,labels=y,colors=c,startangle=90,shadow=True)
plt.legend(loc=0)
plt.title("Payment modes",fontsize='20',fontweight='bold')


plt.subplot(3,2,4)
x = Data['Customer_type'].value_counts()
y = Data['Customer_type'].unique()
c = ['grey','pink']
plt.pie(x,labels=y,colors=c,startangle=90,shadow=True)
plt.legend(loc=1)
plt.title("Customer type",fontsize='20',fontweight='bold')
plt.tight_layout()
````

````python
#Heat map
plt.figure(figsize=(15,7))
plt.style.use('bmh')
sns.set(font_scale=2)
sns.heatmap(Data.corr(),cmap='RdYlBu_r',annot=True)
sns.set(font_scale=2)
plt.xticks(rotation=90,fontsize='18')
[i.set_color("Black") for i in plt.gca().get_xticklabels()]
plt.yticks(fontsize='18')
[i.set_color("Black") for i in plt.gca().get_yticklabels()]
plt.suptitle('Correlational Heatmap',fontsize='20',fontweight='bold')
````
